```python
def plot_bars(df, x, y, hue):
    g = sns.catplot(
        data=df.dropna(), kind="bar",
        x=x, y=y, hue=hue,
        palette="dark", alpha=.6, height=6
    )
    g.despine(left=True)
    g.set_axis_labels("", y)
    g.legend.set_title("")
```

```python
def plot_hist(df, x, hue):
    sns.set_theme(style="ticks")
    g = sns.FacetGrid(df.fillna(0), col=hue, col_wrap=2, height=4)
    g.map(sns.histplot, x, log_scale=False, edgecolor=".3", linewidth=.5)
    plt.show()
```

```python
plot_hist(df, x='age', hue='event')
```

```python
plot_hist(df, x='age', hue='offer_type')
```

- ages are around 20 to 100
- outliers over 120
- we could see ages between 45 to 65 are more attracted to the offers
- specially bogo and discount are the most popular

```python
plot_hist(df, x='income', hue='event')
```

```python
plot_hist(df, x='income', hue='offer_type')
```

- we can see people that have a regular income above 25_000 and 100_000 are generally more attracted to the offers
- a few people get the informational offer but are more uniform
